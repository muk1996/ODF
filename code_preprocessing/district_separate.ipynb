```python
import pandas as pd
```

```python
data = pd.read_csv('../../raw_data/SEC_B_2015.csv', low_memory=False)
data.head()
```

# District Level

```python
district = data['district'].unique().tolist()
```

```python
for x in district:
    data1 = data[data.district == x]
    data1.to_csv(('dist/year_2007/'+ str(x) + '_2007.csv'))
```

# Province Level

```python
province = data['province'].unique().tolist()
```

```python
for x in province:
    data1 = data[data.province == x]
    data1.to_csv(('../../province/year_2015/'+ str(x) + '_2015.csv'))
```
